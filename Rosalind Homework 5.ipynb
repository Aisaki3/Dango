```python
# GRPH (1)

k = 3
f = open('rosalind_grph.txt', 'r')
s = f.read().replace('\n', '')
lst = s.split('>Rosalind_')[1:]

for i in range(len(lst)):
    for j in range(len(lst)):
        if i == j:
            pass
        else:
            if lst[i][-k:] == lst[j][4:4+k]:
                print('Rosalind_' + lst[i][:4] + ' ' + 'Rosalind_' + lst[j][:4])
                    
f.close()
```

```python
# MPRT (2)
# In case of SSL error -> rerun code
# In case of "index of out range" error there is a uniprot database issue -> rerun Rosalind

from Bio import SeqIO
import wget

f = open('Rosalind_mprt.txt', 'r')
lst = f.read().split()

for x in lst:
    prot = 'http://www.uniprot.org/uniprot/{id}.fasta'.format(id = x.split('_')[0])
    wget.download(prot)

print('\n')
for x in lst:
    g = open('{id}.fasta'.format(id = x.split('_')[0]), 'r')
    
    l = []
    for record in SeqIO.parse(g, 'fasta'):
        l.append(record.seq)
    s = str(l[0])
    
    l2 = []
    for i in range(len(s)-3):
        s1 = s[i:i+4]
        if s1[0] == ("N") and s1[1] != "P" and s1[2] in ["S", "T"] and s1[3] != ("P"):
            l2.append(i+1)
    
    
    if l2 != []:
        print(x)
        print(' '.join(map(str, l2)))
    g.close()
                
f.close()
```

```python
# ORF (3)

from Bio import SeqIO

f = open('Rosalind_orf.txt', 'r')

l = []
for record in SeqIO.parse(f, 'fasta'):
    l.append(record.seq)

DNA = str(l[0])
DNAr = ''.join(reversed(DNA))
DNArc = ''
for x in DNAr:
    if x == 'A':
        DNArc += 'T'
    if x == 'C':
        DNArc += 'G'
    if x == 'G':
        DNArc += 'C'
    if x == 'T':
        DNArc += 'A'

DNA = DNA.replace('T', 'U')
DNArc = DNArc.replace('T', 'U')

prot = ['UUU','F','CUU','L','AUU','I','GUU','V','UUC','F',\
        'CUC','L','AUC','I','GUC','V','UUA','L','CUA','L',\
        'AUA','I','GUA','V','UUG','L','CUG','L','AUG','M',\
        'GUG','V','UCU','S','CCU','P','ACU','T','GCU','A',\
        'UCC','S','CCC','P','ACC','T','GCC','A','UCA','S',\
        'CCA','P','ACA','T','GCA','A','UCG','S','CCG','P',\
        'ACG','T','GCG','A','UAU','Y','CAU','H','AAU','N',\
        'GAU','D','UAC','Y','CAC','H','AAC','N','GAC','D',\
        'CAA','Q','AAA','K','GAA','E','CAG','Q','AAG','K',\
        'GAG','E','UGU','C','CGU','R','AGU','S','GGU','G',\
        'UGC','C','CGC','R','AGC','S','GGC','G','CGA','R',\
        'AGA','R','GGA','G','UGG','W','CGG','R','AGG','R',\
        'GGG','G','UAA','Stop','UAG','Stop','UGA','Stop']

prot_dct = dict()
for i in range(0, len(prot), 2):
    prot_dct[prot[i]] = prot[i + 1]

lst = []
for i in range(0,len(DNA)-2,3):
    lst.append(prot_dct[DNA[i]+DNA[i+1]+DNA[i+2]])
s1 = (''.join(lst).rstrip('Stop'))
lst = []
for i in range(1,len(DNA)-2,3):
    lst.append(prot_dct[DNA[i]+DNA[i+1]+DNA[i+2]])
s2 = (''.join(lst).rstrip('Stop'))
lst = []
for i in range(2,len(DNA)-2,3):
    lst.append(prot_dct[DNA[i]+DNA[i+1]+DNA[i+2]])
s3 = (''.join(lst).rstrip('Stop'))
lst = []
for i in range(0,len(DNArc)-2,3):
    lst.append(prot_dct[DNArc[i]+DNArc[i+1]+DNArc[i+2]])
s4 = (''.join(lst).rstrip('Stop'))
lst = []
for i in range(1,len(DNArc)-2,3):
    lst.append(prot_dct[DNArc[i]+DNArc[i+1]+DNArc[i+2]])
s5 = (''.join(lst).rstrip('Stop'))
lst = []
for i in range(2,len(DNArc)-2,3):
    lst.append(prot_dct[DNArc[i]+DNArc[i+1]+DNArc[i+2]])
s6 = (''.join(lst).rstrip('Stop'))

orf = []
start = ''
stop = ''

DNA_dict = {1: s1, 2: s2, 3: s3, 4: s4, 5: s5, 6: s6}
for s in DNA_dict.values():
    for k in range(len(s)):
        if s[k] == 'M':
            start = k
            for j in range(k, len(s)):
                if s[j] == 'S' and s[j+1] == 't':
                    stop = j
                    orf.append(s[start:stop])
                    break
                    
for x in set(orf):
    print(x)

f.close()
```

```python
# PMCH (5)

import math
from collections import Counter
from Bio import SeqIO

f = open('rosalind_pmch.txt', 'r')

lst = []
for record in SeqIO.parse(f, 'fasta'):
    lst.append(record.seq)
s = lst[0]

pmch = math.factorial(Counter(s)['A']) * math.factorial(Counter(s)['C'])
print(pmch)

f.close()
```

```python
# PPER (6)

from math import factorial

f = open('rosalind_pper.txt', 'r')
l = f.read().split()
n, k = int(l[0]), int(l[1])

print(int(factorial(n) / factorial(n - k) % 1000000))
    
f.close()
```

```python
# TREE (7)

import networkx as nx

f = open('rosalind_tree.txt', 'r')
g = open('rosalind_tree.txt', 'rb')

l = f.read().split()
n = int(l[0])

G = nx.read_adjlist(g)
edg = n - G.number_of_edges() - 1
print(edg)

f.close()
g.close()
```

```python
# SSEQ (8)

from Bio import SeqIO
f = open('rosalind_sseq.txt', 'r')

l = []
for record in SeqIO.parse(f, 'fasta'):
    l.append(record.seq)
s = l[0]

s1 = str(l[0])
s2= str(l[1])
str_ind = 0
s1_len = len(s1)

l2 = []
for x in s2:
    str_ind = s1.find(x)
    l2.append(str_ind - len(s1) + s1_len + 1)
    s1 = s1[str_ind+1:]
print(*l2)

f.close()
```

```python
# TRAN (9)

from Bio import SeqIO

f = open('rosalind_tran.txt', 'r')

l = []
for record in SeqIO.parse(f, 'fasta'):
    l.append(record.seq)
s = l[0]

s1 = l[0]
s2 = l[1]
ts = 0
tv = 0

for x,y in zip(s1,s2):
    if x == 'A' and y == 'G':
        ts += 1
    elif x == 'G' and y == 'A':
        ts += 1
    elif x == 'C' and y == 'T':
        ts += 1
    elif x == 'T' and y == 'C':
        ts += 1
    elif x == 'A' and y == 'T':
        tv += 1
    elif x == 'A' and y == 'C':
        tv += 1
    elif x == 'G' and y == 'T':
        tv += 1
    elif x == 'G' and y == 'C':
        tv += 1  
    elif x == 'T' and y == 'A':
        tv += 1
    elif x == 'C' and y == 'A':
        tv += 1
    elif x == 'T' and y == 'G':
        tv += 1
    elif x == 'C' and y == 'G':
        tv += 1  

print(ts / tv)
f.close()
```
