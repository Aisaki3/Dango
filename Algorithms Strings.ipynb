```python
#1 caesarCipher

def caesarCipher(s, k):
    sft = k % 26
    alf = 'abcdefghijklmnopqrstuvwxyz'
    alf_sft = alf[sft:]+alf[:sft]
    alf_sft_u = alf_sft.upper()
    alf_u = alf.upper()
    alf_mix = alf + alf_u

    z = list(zip(alf, alf_sft))
    z2 = list(zip(alf_u, alf_sft_u))
    r = ''

    for x in s:
        if x not in alf_mix:
            r += x
        for i in range(len(z)):
            if x in z[i][0]:
                r += z[i][1]
                break
            elif x in z2[i][0]:
                r += z2[i][1]
                break
    return r
```

```python
#2 Funny string

def funnyString(s):
    s1 = s
    s2 = s[::-1]
    l1,l2,d1,d2 = [],[],[],[]

    for x in s1:
        l1.append(ord(x))
    for k in s2:
        l2.append(ord(k))
        
    for i in range(1,len(l1)):
        d1.append(abs(l1[i]-l1[i-1]))
    for j in range(1,len(l2)):
        d2.append(abs(l2[j]-l2[j-1]))
        
    if d1 == d2:
        return 'Funny'
    else:
        return 'Not Funny'
```

```python
#3 the Love Letter Mystery

def theLoveLetterMystery(s):
    alf = 'abcdefghijklmnopqrstuvwxyz'
    steps = 0
    a = len(s) // 2

    if len(s) % 2 == 0:
        for i in range(0,a):
            c = 0
            while c < 26:
                if s[a-i-1] == s[a+i]:
                    c = 26
                elif alf.index(s[a-i-1]) > alf.index(s[a+i]):
                    s1 = s[:a-i-1]
                    if alf.index(s[a-i-1])-1 < 0:
                        s2 = alf[0]
                    else:
                        s2 = alf[alf.index(s[a-i-1])-1]
                    s3 = s[a-i:]
                    s = s1 + s2 + s3
                    c += 1
                    steps += 1
                elif alf.index(s[a-i-1]) < alf.index(s[a+i]):
                    s1 = s[:a+i]
                    s2 = alf[alf.index(s[a+i])-1]
                    s3 = s[a+i+1:]
                    s = s1 + s2 + s3
                    c += 1
                    steps += 1
                else:
                    pass

    else:
        for i in range(0,a):
            c = 0
            while c < 26:
                if s[a-i-1] == s[a+i+1]:
                    c = 26
                elif alf.index(s[a-i-1]) > alf.index(s[a+i+1]):
                    s1 = s[:a-i-1]
                    s2 = alf[alf.index(s[a-i-1])-1]
                    s3 = s[a-i:]
                    s = s1 + s2 + s3
                    c += 1
                    steps += 1
                elif alf.index(s[a-i-1]) < alf.index(s[a+i+1]):
                    s1 = s[:a+i+1]
                    s2 = alf[alf.index(s[a+i+1])-1]
                    s3 = s[a+i+2:]
                    s = s1 + s2 + s3
                    c += 1
                    steps += 1
                else:
                    pass

    return steps
```

```python
#4 Game of Thrones

def gameOfThrones(s):
    alf = 'abcdefghijklmnopqrstuvwxyz'
    alf_s = set(alf)
    l = list(s)
    l1 = []
    l2 = []

    for x in l:
        if set(l1).union(set(l2)) == alf_s:
            break
            
        if l.count(x) % 2 == 0:
            l2.append(x)
        else:
            l1.append(x)

    l1s = set(l1)
    if len(l1s) > 1:
        return('NO')
    else:
        return('YES')
```

```python
#5 Gemstones

def gemstones(arr):
    gem = []
    c = 0

    for x in arr:
        for j in x:
            c = 0
            while c < len(arr):
                if j in arr[c]:
                    c += 1
                else:
                    c = len(arr)+1
                if c == len(arr):
                    gem.append(j)
    return(len(set(gem)))
```
