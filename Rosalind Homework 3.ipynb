```python
#1 DNA

f = open('rosalind_dna.txt', 'r')

dna = f.read()
lst = []
lst.append(str(dna.count('A')))
lst.append(str(dna.count('C')))
lst.append(str(dna.count('G')))
lst.append(str(dna.count('T')))
print(' '.join(lst))

f.close()
```

```python
#3 REVC

f = open('rosalind_revc.txt', 'r')

DNA = ''.join(reversed(f.read()))
DNAc = ''
for x in DNA:
    if x == 'A':
        DNAc += 'T'
    if x == 'C':
        DNAc += 'G'
    if x == 'G':
        DNAc += 'C'
    if x == 'T':
        DNAc += 'A'
print(DNAc)

f.close()
```

```python
#4 IPRB

f = open('rosalind_iprb.txt', 'r')

lst = list(map(int, f.read().split()))
k = lst[0]
m = lst[1]
n = lst[2]

a = 'HH'
b = 'Hr'
c = 'rr'
g0 = []
g1 = []

for i in range(k):
    g0.append(a)
for i in range(m):
    g0.append(b)
for i in range(n):
    g0.append(c)

for i in range(len(g0)):
    for j in range(len(g0)):
        if j == i:
            pass
        else:
            g1.append(g0[i][0]+g0[j][0])
            g1.append(g0[i][0]+g0[j][1])
            g1.append(g0[i][1]+g0[j][0])
            g1.append(g0[i][1]+g0[j][1])

alleles = 0
for x in g1:
    if 'H' in x:
        alleles +=1

print(alleles/len(g1))

f.close()
```

```python
#5 GC

from Bio import SeqIO

f = open('rosalind_gc.txt', 'r')
    
lst = []
lst_id = []

for x in SeqIO.parse(f, 'fasta'):
    lst.append(x.seq)
    lst_id.append(x.id)
    
p = 0

for x in lst:
        A = x.count('A')
        C = x.count('C')
        G = x.count('G')
        T = x.count('T')

        if (C + G) / (C + G + A + T) > p / 100:
            p = (C + G) / (C + G + A + T) * 100
            ID = lst_id[lst.index(x)]
    
print(ID)
print("%.6f" % p)

f.close()
```

```python
#6 SUBS

f = open('rosalind_subs.txt', 'r')

dna = f.read().split()
lst = []

for i in range(len(dna[0])-len(dna[1])):
    if dna[1] == dna[0][i:i+(len(dna[1]))]:
        lst.append(i+1)

print(' '.join(str(x) for x in lst))

f.close()
```

```python
#7 LIA

from math import factorial as mf

f = open('rosalind_lia.txt', 'r')

lst = list(map(int,f.read().split()))

k = lst[0]
N = lst[1]
    
p = 1/4
pf = 0
ksq = 2**k

for i in range(N,(ksq+1)):
    pf += ((mf(ksq)/math.factorial(i)/mf(ksq-i))*(p**i)*((1-p)**(ksq-i)))

print(pf)

f.close()
```

```python
#8 IEV

f = open('rosalind_iev.txt', 'r')

l = list(map(int,f.read().split()))

val = [1,1,1,0.75,0.5,0]
a = 0

for i in range(len(l)):
    a += l[i]*val[i]*2

print(a)

f.close()
```

```python
#9 CONS

from Bio import SeqIO
from collections import Counter
import pandas as pd

lst = []
for x in SeqIO.parse('rosalind_cons.txt', 'fasta'):
    lst.append(x.seq)

lst_cnt = []
for pos in range(len(lst[0])):
    new_lst = []
    for seq in lst:
        new_lst.append(seq[pos])
        
    lst_cnt.append(Counter(new_lst))

for pos in range(len(lst_cnt)):
    print(lst_cnt[pos].most_common()[0][0], end='')

print()

dft = pd.DataFrame(lst_cnt, columns=['A', 'C', 'G', 'T']).fillna(0).T
for symbol in ['A', 'C', 'G', 'T']:
    print(symbol, end=': ')
    dft.loc[symbol].apply(lambda row: print(int(row), end=' '))
    print()
```

```python
#10 PROB

import math

f = open('rosalind_prob.txt', 'r')

l = f.read().split()
s, A = l[0], list(map(float,l[1:]))
B = []

for x in A:
    temp = 0
    for y in s:
        if y in 'CG':
            temp += math.log(x/2,10)
        else:
            temp += math.log((1-x)/2,10)
    B.append(temp)

B1 = []
for x in B:
    B1.append("%.3f" % x)

print(' '.join(B1))

f.close()
```
