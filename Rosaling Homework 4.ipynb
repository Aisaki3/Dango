```python
# FIB
f = open('rosalind_fib.txt', 'r')
lst = f.read().split()
N = int(lst[0])
k = int(lst[1])

rabbits = 1
rabbits_p = 1

if N == 1:
    print(rabbits_p)
elif N == 2:
    print(rabbits_p)
else:
    for i in range(3,N+1):
        rabbits_y = rabbits_p * k
        rabbits_p = rabbits
        rabbits += rabbits_y
    print(rabbits)

f.close()
```

```python
# HAMM 
f = open('rosalind_hamm.txt', 'r')
lst = f.read().split()
lst1 = lst[0]
lst2 = lst[1]

hamm = 0
for i in range(len(lst1)):
    if lst1[i] != lst2[i]:
        hamm +=1

print(hamm)
f.close()
```

```python
# FIBD

f = open('rosalind_fibd.txt', 'r')
lst = f.read().split()
N = int(lst[0])
m = int(lst[1])

rabbits = 1
rabbits_p = 1
lst1 = [1,0]

for i in range(3,N+1):
    rabbits_y = rabbits_p
    rabbits_p = rabbits
    rabbits += rabbits_y
    lst1.append(rabbits_y)

    if i - m > 0:
        rabbits_p -= lst1[0]
        rabbits -= lst1[0]
        del lst1[0]

print(rabbits)

f.close()
```

```python
# MRNA

f = open('rosalind_mrna.txt', 'r')
s = str(f.read())

prot = ['UUU','F','CUU','L','AUU','I','GUU','V','UUC','F',\
        'CUC','L','AUC','I','GUC','V','UUA','L','CUA','L',\
        'AUA','I','GUA','V','UUG','L','CUG','L','AUG','M',\
        'GUG','V','UCU','S','CCU','P','ACU','T','GCU','A',\
        'UCC','S','CCC','P','ACC','T','GCC','A','UCA','S',\
        'CCA','P','ACA','T','GCA','A','UCG','S','CCG','P',\
        'ACG','T','GCG','A','UAU','Y','CAU','H','AAU','N',\
        'GAU','D','UAC','Y','CAC','H','AAC','N','GAC','D',\
        'CAA','Q','AAA','K','GAA','E','CAG','Q','AAG','K',\
        'GAG','E','UGU','C','CGU','R','AGU','S','GGU','G',\
        'UGC','C','CGC','R','AGC','S','GGC','G','CGA','R',\
        'AGA','R','GGA','G','UGG','W','CGG','R','AGG','R',\
        'GGG','G','UAA','Stop','UAG','Stop','UGA','Stop']

mrna = dict()
for i in range(0,len(prot),2):
    if prot[i+1] in mrna:
        mrna[prot[i+1]].append(prot[i])
    else:
        mrna[prot[i+1]] = [prot[i]]

mrna_len = dict()
for key, value in mrna.items():
    mrna_len[key] = len([item for item in value if item])

counter_stop = 0
for x in prot:
    if x == 'Stop':
        counter_stop += 1

counter = 1
for x in s:
    if x != '\n':
        counter *= mrna_len[x]

print(counter % 1000000 * counter_stop)

f.close()
```

```python
# LCSM

from Bio import SeqIO
f = open('rosalind_lcsm.txt', 'r')
lst = []
for x in SeqIO.parse(f, 'fasta'):
    lst.append(x.seq)
    
s1 = lst[0]
s2 = lst[1]
s3 = lst[2]

mot = ''

for i in range(0,len(s1)):
    for j in range(0,i):
        if s1[j:len(s1)-i+1] in s2 and s1[j:len(s1)-i+1] in s3:
            if len(s1[j:len(s1)-i+1]) > len(mot):
                mot = s1[j:len(s1)-i+1]

        if s1[i:len(s1)-j+1] in s2 and s1[i:len(s1)-j+1] in s3:
            if len(s1[i:len(s1)-j+1]) > len(mot):
                mot = s1[i:len(s1)-j+1]

print(mot)
```

```python
# PERM

from itertools import permutations
import pandas as pd

f = open('rosalind_perm.txt', 'r')
n = int(f.read())

lst = []
for i in range(1,n+1):
    lst.append(i)

a = list(permutations(lst))
df = pd.DataFrame(a)

print(len(df))
print(df.to_string(header = False, index=False))

f.close()
```

```python
# REVP

from Bio import SeqIO
import pandas as pd

f = open('rosalind_revp.txt', 'r')
s = str(SeqIO.read(f, 'fasta').seq)
base = {'A':'T','T':'A','C':'G','G':'C'}

lst1 = []
lst2 = []

for i in range(1,len(s)-2):
    j = 1
    while j < 6:
        if i-j < 0:
            break
        if i+j+2 > len(s):
            break
        if s[i] != base[s[i+1]]:
            break
        if s[i-j] != base[s[i+1+j]]:
            break
        else:
            lst1.append(i-j+1)
            lst2.append(len(s[i-j:i+2+j]))
        j+=1

df = pd.DataFrame(columns = ['index', 'lenght'])
df['index'] = lst1
df['lenght'] = lst2
print(df.to_string(header = False, index=False))

f.close()
```

```python
# Rosalind / LEXF

import itertools
import pandas as pd

f = open('rosalind_lexf.txt', 'r')
lst = f.read().split()
n = int(lst[-1])
del(lst[-1])

a = list(itertools.product(lst, repeat=n))
df = pd.DataFrame(a)
df_sorted = df.sort_values(by=[0,1])
print(df_sorted.to_string(header = False, index=False).replace(' ',''))

f.close()
```
